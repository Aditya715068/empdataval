```python
df4=pd.read_table("Emp4.txt",delimiter=",")
df4.head()
```

```python
df4.columns
```

```python
df4=df4.iloc[0:34,:]
```

```python
sns.barplot(x=df4["States/ UTs"],y=df4["Central fund released - 2014-15"],data=df4)
plt.xticks(rotation=90)
plt.title("Central fund released - 2014-15 in each State/Uts",fontsize=15)
plt.show()
```

```python
fraud_max1= df4['Central fund released - 2014-15'].idxmax()
f1=df4.loc[fraud_max1][1:3]
f1
```

```python
sns.barplot(x=df4["States/ UTs"],y=df4["Central fund released - 2015-16"],data=df4)
plt.xticks(rotation=90)
plt.title("Central fund released - 2015-16 in each State/Uts",fontsize=15)
plt.show()
```

```python
fraud_max= df4['Central fund released - 2015-16'].idxmax()
f2=df4.loc[fraud_max][1:4:2]
f2
```

```python
sns.barplot(x=df4["States/ UTs"],y=df4["Central fund released - 2016-17*"],data=df4)
plt.xticks(rotation=90)
plt.title("Central fund released - 2016-17 in each State/Uts",fontsize=15)
plt.show()
```

```python
fraud_max= df4['Central fund released - 2016-17*'].idxmax()
f3=df4.loc[fraud_max][1:4:2]
f3
```

```python
result=pd.DataFrame({"Central fund released Year wise":["2014-15","2015-16","2016-17"],"Central fund released States":[f1[0],f2[0],f3[0]],"Fraud":[f1[1],f2[1],f3[1]]})
result
```

```python
sns.barplot(x=result["Central fund released States"],y=result["Fraud"],hue=result["Central fund released Year wise"],data=result)
plt.show()
```
